```python
median = df.loc[df['flipper_length_mm']<200, 'flipper_length_mm'].median()
df.loc[df['flipper_length_mm'] > 200, 'flipper_length_mm'] = np.nan
df.fillna(median,inplace=True)
```

Correlation of independent layers

```python
target_variable = 'body_mass_g'
independent_variables = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm']
```

```python
correlation_matrix = df[independent_variables + [target_variable]].corr()
correlation_with_target = correlation_matrix[target_variable].drop(target_variable)
```

```python
print("Correlation with the target variable:")
print(correlation_with_target)
```

Categorial columns

```python
categorical_columns = df.select_dtypes(include=['object']).columns

for column in categorical_columns:
    if len(df[column].unique()) <= 200:
        df = pd.get_dummies(df, columns=[column], prefix=[column])
    else:
        df[column] = df[column].astype('category').cat.codes

categorical_columns
```

Dependent and Independent variables

```python
# Independent variables (features)
X = df.drop('flipper_length_mm', axis=1)
X
```

```python
# Dependent variable (target)
y = df['flipper_length_mm']
y
```

Scaling data

```python
#min-max scaling
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
scaled_df
```

```python
# Z-scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
scaled_df
```
